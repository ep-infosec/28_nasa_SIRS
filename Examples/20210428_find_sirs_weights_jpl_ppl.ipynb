## 1. Introduction

Roman data from the JPL Precision Projector Lab (PPL) have some correlated noise. The purpose of this notebook is to see if SIRS can reduce it. This notebook is in Julia 1.5.3. There exists a python back-end that can be used to apply its results to FITS files using python-3. To keep it simple, we're just working with Julia until we know that SIRS helps.

In this notebook, we compute the SIRS frequency dependent weights. This is the time consuming part. The python-3 notebook 20210420_sirs_ppl.ipynb completes the demonstration by applying he weights that are computed here to the data.

Note that the type of some cells is set to "Raw". This is done to prevent those cells from executing. To execute them, change the type to "Code".

```julia
# Standard imports
using Revise        # Handy for interactive work
using Glob          # Filename globbing
using FITSIO        # Julia's standard FITS package
using Jacobi        # Legendre polynomials defined here
using LinearAlgebra # Moore-Penrose inverse is here
using OMEinsum      # Tensor operations
using StatsBase     # Robuts statistics
using Plots         # Plotting

# Simple Improved Reference Subtraction (SIRS)
using SIRS
```

```julia
# The data are here on ADAPT
ddir = "/att/gpfsfs/briskfs01/ppl/cshapiro/JPL-PPL/H4RG-18241/2021-04-15"

# Put the results here
rdir = "/att/gpfsfs/briskfs01/ppl/brausche/roman_caldir/JPL_PPL_Results/";
```

```julia
# Get a list of the input files
files = glob("*.fits", ddir);
files = files[1:5]; # STUB - Do just a few to speed things up
```

```julia
# Instantiate a SIRSCore. From looking at the FITS headers, I know that
# all files have 60 frames sampled up-the-ramp.
#
# According to Chaz Shapiro, the new row overhead (nroh)
# is different from in DCL data. According to Chaz, it is 26.48 µs.
# SIRS assumes that it is an integral number of pixel readout times.
# We will see how that imprints today. For now, say nroh = 5 pixels.
#
# PPL parameters as follows.
hxrg_kind = "h4rg" # Type of HxRG
nout      = 32     # Number of outputs
nroh      = 5      # See above
τ         = 5.e-6  # 200 kHz clocking
naxis3    = 58     # We will be discarding the first 2 of 60 frames per Chaz
sc = SIRSCore(hxrg_kind, nout, nroh, τ, naxis3); # Discard 1st two frames!
```

```julia
# Solve for the frequency dependent weights.
#
# This cell can be very time consuming. The execution time on
# Prism with 40 cores enabled is about 65 seconds/file.

for file in files
    
    # Show some status
    println("Processing file: ", file)
   
    # These are JPL PPL data. The format is a single datacube.
    # Like in the DCL, charge seems to integrate down. For now,
    # just leave this. 
    f = FITS(file, "r")
    D = Float64.(read(f[1]))
    close(f)
    
    # Consider dropping frame #1 before coadd. Chaz says that the
    # system is unstable.
    #
    # * First saved run on 4/19 kept 1st frame
    # * Second saved run on 4/19 discards it and the second,
    #   Which I found was often being thrown out.
    D = D[:,:,3:end]
    
    # Coadd
    coadd!(sc, D)

end
```

```julia
# Solve for the frequency dependent weights, alpha and beta
solve!(sc);
```

```julia
# Save results
output_filename = rdir*"2021-05-25T14:24:00_PPL_18241_95.0K_-1.0V_SIRS.jld"
SIRS.save(sc, output_filename)
```
