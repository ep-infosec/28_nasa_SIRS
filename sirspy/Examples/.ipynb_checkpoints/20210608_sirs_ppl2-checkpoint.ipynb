## 4. Demonstration that SIRS is Beneficial

To show SIRS's effect, fit and save integrated counts images, where counts = slope x time. I typically fit Legendre polynomials, so

<center>$\textrm{counts} = 2 g_c \lambda_1,\hspace{18pt}(1)$</center>

where $g_c$ is the conversion gain in $e^-/\textrm{DN}$.

```python
# Instantiate a Legendre fitting object
legfit = sirspy.LegFit(59,1) # 59 UTR samples, 1st degree fit.
                             # Chaz says to discard frame zero.
```

```python
def fit_many(files, sirs=None):
    """
    Fit many FITS files to compute integrated charge
    
    Parameters: files, list
                  List of input files
                sirs, SIRS object
                  If set, use SIRS reference correction. Otherwise,
                  use legacy rowsonly.
    """
    global gc     # Conversion gain
    global pinv_L # Legendre fitting matrix
    offset = 128  # Offset images by this much and save as uint16
                  #   This is done to minimize downloading time when
                  #   telecommuting.
    result = np.empty((len(files),4096,4096), dtype=np.uint16)

    idx = 0 # A counter
    for file in files:
        
        # Show status
        print('Processing file: ', file)
        
        # Get data
        with fits.open(file) as hdul:
            D = -np.float64(hdul[0].data)[1:] # Integrate up and drop 1st read
        
        # Reference correct
        if sirs is None:
            # Legacy rowsonly reference correction
            sirspy.rowsonly(D) 
        else:
            # SIRS. This particular detector has some bad reference pixels
            # in the right hand reference columns. Setting pplfix1=True
            # linearly interpolates over them.
            sirs.refcor(D, pplfix1=True)
            
        # Fit and compute integrated charge in e-
        Q = np.round(2 * gc * legfit.fit(D)[1])
        
        # Convert to uint16
        Q += offset           # Offset
        Q[Q<0] = 0            # Gracefully handle bad pixels
        Q[Q>2**16-1] = 2**16-1  # ...
        result[idx] = np.array(Q, dtype=np.uint16)       # Save result
        
        idx += 1 # Increment counter        
        
    # Done!
    return(result)
```

```python
# Rowonly correction
result = fit_many(files[:nproc])
fits.PrimaryHDU(result).writeto(rdir+'rows_only.fits', overwrite=True)
```

```python
# SIRS
result = fit_many(files[:nproc], sirs=sirs)
fits.PrimaryHDU(result).writeto(rdir+'sirs.fits', overwrite=True)
```

## 5. Summary

The intial attempt with JPL PPL data shows that SIRS is very beneficial. It visibly reduces the $1/f$ noise. It visibly reduces banding.

There is however room for improvement. SIRS would work somewhat better if the time overhead between ending one row and starting the next were to be an integral number of pixel readout times.
